```python
import pandas as pd
import numpy as np
```

```python
df_customers = pd.read_csv("customer_mix_sample.csv")
```

```python
def create_client_mix_dataframe(df_customers, multiplier_values, product_names):

    df_clientmix =  pd.DataFrame({"CustomerID": df_customers['CustomerID'].unique()})
    
    for product in product_names:
        df_clientmix = pd.merge(df_clientmix, pd.DataFrame({"CustomerID":df_customers[df_customers.Product == product].CustomerID.unique(), product: 1}), \
                                         how='left', on='CustomerID')
        # checking
        if df_clientmix[product].sum() != df_customers[df_customers.Product == product].CustomerID.nunique():
            print(0/0)
    
        df_clientmix[product] *= multiplier_values[product]
    
    df_clientmix['Tag'] = df_clientmix[df_clientmix.columns[1:]].sum(axis=1)    
    
    return df_clientmix

def get_productmix_names_productcounts(location_values, number):
    binary_string = str(bin(number))[2:][::-1]
    products = [location_values[x] for x, v in enumerate(binary_string) if v == '1']
    return {"ProductNames": " and ".join(products), "ProductCount": str(len(products))}
```

```python
def get_product_mix(df_customers):
    product_names = df_customers.Product.unique()

    multiplier_values = {product_names[i]:2**i for i in range(len(product_names))}
    location_values = {i:product_names[i] for i in range(len(product_names))}
    
    df_clientmix2 = create_client_mix_dataframe(df_customers, multiplier_values, product_names)

    product_mix_df = df_clientmix2[['Tag']].groupby('Tag').agg('value_counts').reset_index()
    product_mix_df.columns = ['Tag', 'ClientCount']
    product_mix_df[['ProductNames', 'ProductCount']] = product_mix_df.apply(lambda row: get_productmix_names_productcounts(location_values, int(row.Tag)), axis='columns', \
                                                                  result_type='expand')
    return product_mix_df[['ProductNames','ClientCount', "ProductCount"]].sort_values(by=['ProductCount', 'ClientCount'], ascending=[True, False]).reset_index(drop=True)
```

```python
get_product_mix(df_customers)
```
